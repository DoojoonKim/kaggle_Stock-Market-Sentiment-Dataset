# Text 데이터 처리해보기

```python
word_list = [] # word를 담을 리스트를 만든다.
for i in range(len(df)): # df의 length를 for문의 최대 count로 설정
    x = df.Text[i].split() # ' '을 기준으로 Text 컬럼내의 value를 split한다.
    for y in x: # split을 하면 하나의 list에 키워드들이 담기는데, 
                # 여기서 리스트의 length를 기준으로 하고 
                # 각각의 값들을 불러온다.
        word_list.append(y) # 각각의 값들을 word_list에 append해준다.
```

```python
word_list
```

```python
from collections import Counter
```

```python
word_counter = Counter(word_list) # 어떤 데이터가 count가 몇 번 됐는지를 
                                  # 한 번에 보여주는 함수
word_counter
```

```python
counted_word_dict = dict(word_counter) # word_count를 dict타입으로 바꿔준다.
counted_word_dict
```

```python
# dict타입의 데이터 key, values, items에 대해 다시 회고 할겸 데이터들이 어떻게
# 나오는지 확인하기
display(counted_word_dict.keys())
display(counted_word_dict.values())
display(counted_word_dict.items())
```

```python
sorted_word_dict = sorted(counted_word_dict.items(), key = lambda x: x[1], reverse = True)
```

```python
sorted_word_dict[:20]
```

```python
num = 100
list1 = list()
list2 = list()
for i in range(num):
    list1.append(word_counter.most_common(num)[i][0])
    list2.append(word_counter.most_common(num)[i][1])
```

```python
list2[:10] # 제대로 됐는지 확인
```

```python
import plotly.graph_objects as go
from plotly.offline import iplot
import plotly.express as px
# 모든 Plotly Express 함수는 제공된 인수에 따라 
# 데이터 및 레이아웃이 미리 채워진 graph_objects.Figure 객체를 반환합니다.

fig = px.bar(df, x = df.Sentiment.unique(), # 1, -1을 기준으로 x축을 삼자
             y = df.Sentiment.value_counts(), # 1, -1가 count된 값을 y축으로 하자
             color = ['1', '-1'],
             text = df.Sentiment.value_counts())
fig.update_traces(hovertemplate="Sentiment:'%{x}' Counted: %{y}")
fig.update_layout(title={"text":"Sentiment Counts"},xaxis={"title":"Sentiment"},yaxis={"title":"Count"})
fig.show()
```

```python
fig2 = px.bar(x = list1, y = list2,
              color = list2, hover_name = list1)
fig2.update_traces(hovertemplate="Word:'%{x}' Counted: %{y}")
fig2.update_layout(title={"text":"Word Counts"},xaxis={"title":"Words"},yaxis={"title":"Count"})
fig2.show()
```
